# Evaluation Metrics

Confusion Matrix

```python
class ConfusionMatrix:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.matrix = [[0] * num_classes for _ in range(num_classes)]

    def update(self, true_label, predicted_label):
        self.matrix[true_label][predicted_label] += 1

    def get_matrix(self):
        return self.matrix
```

Accuracy

```python
class Accuracy:
    def __init__(self):
        self.correct = 0
        self.total = 0

    def update(self, true_label, predicted_label):
        if true_label == predicted_label:
            self.correct += 1
        self.total += 1

    def get_accuracy(self):
        return self.correct / self.total if self.total > 0 else 0.0
```

Precision

```python
class Precision:
    def __init__(self, num_classes):
        self.num_classes = num_classes
        self.true_positives = [0] * num_classes
        self.false_positives = [0] * num_classes

    def update(self, true_label, predicted_label):
        if predicted_label == true_label:
            self.true_positives[predicted_label] += 1
        else:
            self.false_positives[predicted_label] += 1

    def get_precision(self):
        precision_scores = []
        for i in range(self.num_classes):
            tp = self.true_positives[i]
            fp = self.false_positives[i]
            precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
            precision_scores.append(precision)
        return precision_scores
```
